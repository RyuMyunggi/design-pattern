## Anti Design Pattern

## 안티 패턴

* 일반적인 디자인 패턴들과는 다르게 소프트웨어 개발자로서 해서는 안되는 행동을 나타내는 것
* 소프트웨어 디자인 원칙은 개발자가 설계 단계에서 지켜야 할 몇 가지 원칙과 가이드라인을 제시함. 로버트 마틴에 의하면 잘못된 설계에는 4가지 특성이 있음
  * 유연하지 않다: 재사용이 어렵다
  * 융통성이 없다: 간단한 수정도 여러 부분을 손봐야함
  * 취약하다: 수정할수록 시스템이 취약해짐
  * 점성이 높다: 설계상의 결함에 대한 수정이 어려워 겉돌게 됨

* 안티패턴은 이러한 문제들에 대한 비효율적이고 비생산적인 해결책임

## 안티 패턴의 원인
* 개발자의 소프트웨어 개발 방식이 미숙
* 개발자가 상황에 맞지 않는 디자인 패턴을 사용

## 안티 패턴의 분류
### 소프트웨어 개발 안티 패턴
* 프로젝트 개발을 시작할 때 코드 구조부터 설계를 함
* 개발이 진행될수록 최초의 코드 구조에서 벗어나게 되는 원인
  * 개발이 진행되면서 개발자의 사고력이 높아짐
  * 고객의 피드백에 따라 사용 패턴이 바뀜
  * 기능 추가 및 확장성을 고려해 초기에 구상한 자료 구조가 수정
* 스파게티 코드
  * 가장 흔하게 접하는 안티 패턴
  * 소프트웨어도 충동적으로 개발하다 보면 흐름이 뒤죽박죽 섞임. 이러한 스파게티 코드는 유지보수와 최적화가 어렵다
  * 스파게티 코드의 원인
    * 객체지향 프로그래밍에 및 분석에 대한 이해 부족
    * 깊게 생각하지 않은 제품 설계 및 디자인
    * 대충 빨리 고치려는 성격
  * 황금망치
    * 적합하지 않은 상황에도 한 번 사용해본 솔루션을 반복적으로 사용하려는 경향이 있음
    * 어떤 못도 박을 수 있는 황금 망치라는 솔루션을 모든 것에 적용하려함
    * 황금망치가 생기는 원인
      * 주어진 문제와 환경에 대해서 잘 알지 못하는 외부인이 관여하는 경우
      * 전혀 다른 목적의 프로젝트에서 성공적으로 적용된 솔루션을 재사용하는 경우
      * 회사의 자원으로 이미 직원들이 잘 알고 교육을 받은 솔루션을 지속적으로 사용하는 경우
    * 황금 망치의 결과
      * 특정 솔루션이 모든 프로젝트에서 사용됨
      * 제품의 기능이 아닌 개발에 사용된 기술 위주로 설명함
      * 개발자들이 '저 솔루션을 사용했어야 됐는데' 라는 탄식을 자주함
      * 사용자의 요구가 충족되지 않음
  * 용암류
    * 프로그램이 망가질까봐 두려워서 건드리지 못하는 죽은 코드나 쓸 수 없는 코드를 일컫는 말
    * 용얌류가 생기는 원인
      * 테스트와 에러 코드가 지나치게 많은 경우
      * 리뷰 없이 단독적으로 코드를 작성하고 인수인계 없이 다른 팀에 넘기는 경우
      * 코드를 이해하는 사람이 아무도 없는 경우
    * 용얌류의 증상
      * 테스트 코드의 범위가 좁다
      * 알 수 없는 주석된 코드가 많음
      * 쓰이지 않는 인터페이스가 생기거나 기존 코드를 우회하는 방식으로 개발
  * 복 붙 프로그래밍
    * 가장 흔한 안티 패턴
    * 복붙이 발생하는 원인
      * 코딩 및 개발이 미숙한 초보 개발자인 경우
      * 빠르게 버그를 수정하고 넘어가야하는 경우
      * 모듈 간의 표준화 또는 구조 단일화를 위한 코드 중복이 생길 때
      * 장기적인 사고의 부재
    * 복붙의 결과
      * 유사한 문제가 애플리케이션의 여러 부분에서 발생
      * 높은 유지보수 비용과 버그 발생
      * 코드 중복으로 인해 모듈식 코드가 줄어듦
      * 동일한 몬제가 계속 발생

### 소프트웨어 설계 안티 패턴
* 소프트 웨어 설계는 개발 및 테스트팀과 PM등의 관계자들이 함께 모델링 함
* 기초 설계는 제품의 성공을 좌지우지함
* 시간 낭비
  * 설계의 재사용을 의미
  * 시간낭비의 원인
    * 문서 또는 설계 단계의 문제점 및 해결책 공유 부재
    * 집단 내 소통 부족
    * 처음부터 새롭게 개발하는 방식이 관습인 집단. 개발 프로세스 및 규칙이 없는 경우
  * 시간 낭비의 결과
    * 간단한 문제에 대한 쓸데 없는 논의
    * 시간과 리소스의 증가로 인한 예산 낭비와 늦춰지는 출시일
    * 한 제품에만 적용되는 폐쇄적인 설계와 노력의 중복, 부실한 위기 관리
* 제품/기술 종속
  * 특정 기술이 시스템에 종속돼 떼어낼 수가 없는 구조가 돼버렸을 때
  * 제품/종속이 발생하는 원인
    * 벤더사와의 유착 관계 또는 사용 요금 할인
    * 기술 및 성능이 아닌 마케팅과 세일즈 관점에서 선택한 제품
    * 수익적으로 성공한 다른 프로젝트에서 쓰인 제품을 목적 및 요구 사항이 다른 프로젝트에 그대로 접목하는 경우
    * 기술자나 개발자가 이미 익숙한 제품 선택
  * 제품/기술 종속의 결과
    * 제품 출시 및 유지 보수 주기를 벤더사의 제품의 출시 시기에 맞춤
    * 고객의 요구 보다는 제품의 기술 중심으로 개발
    * 출시일이 불확실하고 고객의 기대치에 미치지 못함
* 다수 디자인
  * 조직의 프로세스에 따라 여러 사람이 모여 함계 시스템을 설계 하는 경우가 있음
  * 다수 디자인의 원인
    * 많은 관계자들의 의견을 수렴하는 분위기의 조직인 경우
    * 설계를 맡은 총책임자가 없는 경우
    * 고객의 요구가 아닌 마케터나 기술자의 의견이 우선시 될 때
  * 다수 디자인 안티 패턴의 결과
    * 설계가 끝난 뒤에도 개발자와 설계자 사이에 의견이 갈림
    * 문서화하기 힘들 정도로 설계가 복잡해짐
    * 수정사항이 여러 사람을 거쳐야 하고 개발 시간이 지연됌
